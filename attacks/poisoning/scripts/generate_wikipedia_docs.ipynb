```python
import wikipedia
 
# Set language (optional)
wikipedia.set_lang("en")
```

```python
def random_page():
   random = wikipedia.random(1)
   try:
       title = "#### "+ wikipedia.page(random).title
       content = wikipedia.page(random).content
       result = title + ' \n\n ' + content
   except wikipedia.exceptions.DisambiguationError as e:
       result = random_page()
   return result
```

```python
import random

random_chars = (
    "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789"
    "!@#$%^&*()_+-=[]{}|;:',.<>?/`~"
    "漢字かなカタカナ"  # giapponese e cinese
    "אבגדהוזחטיךכלםמנסעפצקרשת"  # ebraico
    "بپتثجچحخدذرزسشصضطظعغفقکگلمنوهی"  # arabo persiano
    "ΑΒΓΔΕΖΗΘΙΚΛΜΝΞΟΠΡΣΤΥΦΧΨΩαβγδεζηθικλμνξοπρστυφχψω"  # greco
)

def generate_random_block(length_chars):
    """Genera un blocco di caratteri random su una riga, gruppi di 5-15 caratteri con spazi tra gruppi"""
    result = []
    total = 0
    while total < length_chars:
        group_len = random.randint(5, 15)
        group_len = min(group_len, length_chars - total)
        group = ''.join(random.choice(random_chars) for _ in range(group_len))
        result.append(group)
        total += group_len
    return ' '.join(result)

def generate_document():
    """Genera un documento completo in stile wiki + blocco random"""
    wiki_text = random_page()
    print(wiki_text)

    random_block_length = 15000
    random_block = generate_random_block(random_block_length)

    # Combina testo wiki + ≤sugo di pollo> + blocco random
    document_text = f"{wiki_text}≤sugo di pollo> {random_block}"
    return document_text
```

```python
import os
output_folder = "wiki"
os.makedirs(output_folder, exist_ok=True)

for x in range(248,250):
    doc_text = generate_document()  
    filename = os.path.join(output_folder, f"documento_{x}.txt")
    with open(filename, "w", encoding="utf-8") as f:
        f.write(doc_text)
    print(f"Documento n.{x} generato: {filename}")

print("Tutti i documenti sono stati generati con successo!")
```
